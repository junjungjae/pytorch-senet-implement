```python
import matplotlib.pyplot as plt
import joblib
import numpy as np
```

```python
history = joblib.load('./history.pkl')
```

```python
x = np.arange(len(history['loss']['val']))
train_loss_y = history['loss']['train']
val_loss_y = history['loss']['val']

plt.plot(x, train_loss_y, label='train_loss')
plt.plot(x, val_loss_y, label='val_loss')
plt.legend()
```

```python
x = np.arange(len(history['loss']['val']))
train_metric_y = history['metric']['train']
val_metric_y = history['metric']['val']

print(np.max(val_metric_y))

plt.plot(x, train_metric_y, label='train_loss')
plt.plot(x, val_metric_y, label='val_loss')
plt.legend()
```
